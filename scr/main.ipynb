```python
import yfinance
import pandas as pd
import requests
from bs4 import BeautifulSoup

def fetch_sectors_names():
    url = "https://stockanalysis.com/stocks/industry/sectors/"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        df=pd.read_html(str(soup.find_all("table")))[0]
    else:
        print(f"Error: Failed to fetch data from page {url}")

    return df

def fetch_industry_names():
    url = "https://stockanalysis.com/stocks/industry/all/"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        df=pd.read_html(str(soup.find_all("table")))[0]
    else:
        print(f"Error: Failed to fetch data from page {url}")

    return df

def fetch_data(sectors):
    url = f"https://stockanalysis.com/stocks/sector/{sectors}/"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        df=pd.read_html(str(soup.find_all("table")))[0]
        df.drop(columns='No.', inplace=True)
    else:
        print(f"Error: Failed to fetch data from page {url}")

    return df
```

```python
finance = pd.read_csv('../data/stock_sectors/financials.csv')
technology = pd.read_csv('../data/stock_sectors/technology.csv')
health = pd.read_csv('../data/stock_sectors/healthcare.csv')
```

```python
try:
  data_finance = yfinance.download(list(finance["Symbol"]), start='2005-01-01')
  data_open_finance = data_finance['Open'].resample('W-MON').last().pct_change()+1
except:
  print("error")
```

```python
technology = technology.drop(541) # it's null value
```

```python
technology_symbol_list = technology["Symbol"].tolist()
data_tech = yfinance.download(technology_symbol_list, start='2005-01-01')
data_open_technology = data_tech['Open'].resample('W-MON').last().pct_change()+1
```

```python
health = health.drop(98) # unreachable
health_symbol_list = health["Symbol"].tolist()
data_health = yfinance.download(health_symbol_list, start='2005-01-01')
data_open_health = data_health['Open'].resample('W-MON').last().pct_change()+1


```

```python
data_open_finance.fillna(data_open_finance.mean(), inplace=True)
data_open_health.fillna(data_open_health.mean(), inplace=True)
data_open_technology.fillna(data_open_technology.mean(), inplace=True)
```

```python
df_finance = data_open_finance
df_health = data_open_health
df_technology = data_open_technology

dates_finance = pd.date_range(start='2005-01-01', periods=len(df_finance), freq='W')
dates_health = pd.date_range(start='2005-01-03', periods=len(df_health), freq='W')
dates_technology = pd.date_range(start='2005-01-01', periods=len(df_technology), freq='W')

df_finance['Date'] = dates_finance
df_health['Date'] = dates_health
df_technology['Date'] = dates_technology

df_finance = df_finance[['Date'] + [col for col in df_finance.columns if col != 'Date']]
df_health = df_health[['Date'] + [col for col in df_health.columns if col != 'Date']]
df_technology = df_technology[['Date'] + [col for col in df_technology.columns if col != 'Date']]

melted_df_finance = pd.melt(df_finance, id_vars='Date', var_name='Ticker', value_name='Open')
melted_df_health= pd.melt(df_health, id_vars='Date', var_name='Ticker', value_name='Open')
melted_df_technology = pd.melt(df_technology, id_vars='Date', var_name='Ticker', value_name='Open')
```

```python
melted_df_technology
```

```python
!pip install tsfresh
import tsfresh
```

```python
melted_df_finance = melted_df_finance.dropna()
melted_df_health = melted_df_health.dropna()
melted_df_technology = melted_df_technology.dropna()

df_finance_extracted = tsfresh.extract_features(melted_df_finance, column_id='Ticker', column_sort='Date',
                                       default_fc_parameters=tsfresh.feature_extraction.EfficientFCParameters())

df_technology_extracted = tsfresh.extract_features(melted_df_technology, column_id='Ticker', column_sort='Date',
                                       default_fc_parameters=tsfresh.feature_extraction.EfficientFCParameters())

df_health_extracted = tsfresh.extract_features(melted_df_health, column_id='Ticker', column_sort='Date',
                                       default_fc_parameters=tsfresh.feature_extraction.EfficientFCParameters())


```

```python
df_finance_extracted
```

```python
df_finance_extracted.dropna(thresh=0.80, axis=1, inplace=True)
df_finance_extracted.fillna(df_finance_extracted.mean(), inplace=True)
df_finance_extracted["target"] = "f"

df_health_extracted.dropna(thresh=0.80, axis=1, inplace=True)
df_health_extracted.fillna(df_health_extracted.mean(), inplace=True)
df_health_extracted["target"] = "s"

df_technology_extracted.dropna(thresh=0.80, axis=1, inplace=True)
df_technology_extracted.fillna(df_technology_extracted.mean(), inplace=True)
df_technology_extracted["target"] = "t"
```

```python
df_technology_extracted
```

```python
combined_df = pd.concat([df_finance_extracted, df_health_extracted, df_technology_extracted])
combined_df
```

### Scaling

```python
from sklearn.preprocessing import RobustScaler
scaler_std = RobustScaler()

X = combined_df.drop(columns=['target'])
y = combined_df['target']

scaler_std.fit(X)
X_scaled_std = scaler_std.transform(X)
X_scaled_std = pd.DataFrame(X_scaled_std, columns=X.columns)
X_scaled_std
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_std,y,test_size=0.2, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV


clf = SVC(probability=True)
param_grid = {'C': [0.1, 1, 10], 'gamma': [0.001, 0.01,0.1]}
grid_search_svc = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
grid_search_svc.fit(X_train, y_train)
grid_search_svc
```

```python
classifier_svc = grid_search_svc.best_estimator_
classifier_svc.fit(X_train, y_train)
classifier_svc
```

```python
from sklearn import metrics

predictions = classifier_svc.predict(X_test)

accuracy = metrics.accuracy_score(y_test, predictions)
accuracy
```

```python
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

initial_type = [('float_input', FloatTensorType([None, X_train.shape[1]]))]
onnx_model = convert_sklearn(classifier_svc, initial_types=initial_type)

# save model
with open("best_model_svc2.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

clf = KNeighborsClassifier()
param_grid = {'n_neighbors': np.arange(1, 20, 1)}

grid_search_KNeighbors = GridSearchCV(clf, param_grid, cv=10, scoring="accuracy")
grid_search_KNeighbors = grid_search_KNeighbors.fit(X_train, y_train)
classifier_KNeighbors = grid_search_KNeighbors.best_estimator_
classifier_KNeighbors
```

```python
classifier_KNeighbors.fit(X_train, y_train)
predictions = classifier_KNeighbors.predict(X_test)
accuracy = metrics.accuracy_score(y_test, predictions)
accuracy
```

```python
onnx_model = convert_sklearn(classifier_KNeighbors, initial_types=initial_type)

# Modeli dosyaya kaydetme
with open("best_model_KNN2.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())
```
